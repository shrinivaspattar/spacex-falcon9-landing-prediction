```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0

    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return R * c
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

## Findings and Observations

### 1. Proximity to Equator
The launch sites are located in Florida and California, which are relatively closer to the equator compared to higher latitude regions. This is beneficial because Earth’s rotation provides additional velocity to rockets, improving launch efficiency.

### 2. Proximity to Coastline
All launch sites are very close to the coastline. This is intentional because rockets are launched over the ocean to reduce risk to human life and infrastructure in case of failure.

### 3. Proximity to Cities
The launch sites are not located directly inside major cities. They maintain a safe distance from densely populated areas to minimize risk, but are still close enough for logistical support.

### 4. Proximity to Highways and Railways
Launch sites are connected to nearby highways and infrastructure, which helps in transporting equipment and materials efficiently.

### 5. Distance Observations
From the plotted distance lines:
- The coastline is very close to launch sites (few kilometers)
- Cities are relatively farther away
- Infrastructure (roads, etc.) is accessible but not too close

Launch sites are strategically chosen to balance:
- Safety (far from cities)
